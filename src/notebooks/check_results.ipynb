```python
import xarray as xr
from netCDF4 import Dataset

filepath = "/workspace/results/corrdiff_output.nc"

nc = Dataset(filepath)

print(nc.groups)          # top-level groups
# print(nc.groups["group_name"].variables)  # variables inside a group
```

```python
ds_truth = xr.open_dataset(filepath, group="truth")
ds_prediction = xr.open_dataset(filepath, group="prediction")
ds_input = xr.open_dataset(filepath, group="input")
```

```python
timei = 0
```

```python
X = ds_input.x_t2m.isel(time=timei)
X.plot(figsize=(5, 7))
```

```python
Y = ds_truth.y_t2m.isel(time=timei)
Y.plot(figsize=(5, 7))
```

```python
Y_p = ds_prediction.y_t2m.isel(time=timei)
Y_p.plot(figsize=(5, 7))
```

```python
(X - Y).plot(figsize=(5, 7))
```

```python
(Y_p - Y).plot(figsize=(5, 7))
```
