```python
import sys
sys.path.insert(0, "/workspace/src")

from datasets.carra2 import ZarrDataset

DATA_PATH  = "/workspace/data/input_data_era5tocarra2/data.zarr"
STATS_PATH = "/workspace/data/input_data_era5tocarra2/stats.zarr"

dataset = ZarrDataset(data_path=DATA_PATH, stats_file=STATS_PATH)
print(f"Dataset length: {len(dataset)}")
```

```python
# Variable groups as discovered by ZarrVariableDiscovery (same logic as training)
vg = dataset.variable_groups
print(f"Dynamic inputs  ({len(vg.dynamic_inputs)}): {vg.dynamic_inputs}")
print(f"Static inputs   ({len(vg.static_inputs)}):  {vg.static_inputs}")
print(f"Targets         ({len(vg.targets)}): {vg.targets}")
```

```python
# Normalization stats for every variable (mean and std as loaded during training)
print(f"{'Variable':<30} {'Mean':>12} {'Std':>12}")
print("-" * 56)
for var, mean, std in zip(
    dataset.input_vars,
    dataset.input_means.squeeze().tolist(),
    dataset.input_stds.squeeze().tolist(),
):
    print(f"  {var:<28} {mean:12.4f} {std:12.4f}")

print()
print(f"{'Target':<30} {'Mean':>12} {'Std':>12}")
print("-" * 56)
for var, mean, std in zip(
    dataset.target_vars,
    dataset.target_means.squeeze().tolist(),
    dataset.target_stds.squeeze().tolist(),
):
    print(f"  {var:<28} {mean:12.4f} {std:12.4f}")
```

```python
# Inspect one sample and spatial shape
y, x = dataset[0]
print(f"Input tensor  shape: {tuple(x.shape)}  (channels={x.shape[0]}, H={x.shape[1]}, W={x.shape[2]})")
print(f"Target tensor shape: {tuple(y.shape)}  (channels={y.shape[0]}, H={y.shape[1]}, W={y.shape[2]})")
print(f"\nTime range: {dataset.time()[0]}  →  {dataset.time()[-1]}")
```
